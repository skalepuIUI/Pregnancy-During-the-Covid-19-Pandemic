## Outliers treatment

```python
plotting_columns = df[["Maternal_Age", "Threaten_Baby_Danger", "Threaten_Baby_Harm", "PROMIS_Anxiety"]]

#IQR is being calucalted for each column
Q1 = plotting_columns.quantile(0.25)

Q3 = plotting_columns.quantile(0.75)

Interquartilerange = Q3 - Q1

# to detect outliers, upper and lower bounds are being defined

lower_bound = Q1 - 1.5 * Interquartilerange
upper_bound = Q3 + 1.5 * Interquartilerange

#for each column, outliers are being found
outliers_of_each_column = ((plotting_columns < lower_bound) | (plotting_columns > upper_bound)).any(axis=1)

# To print each column outlier number
print(outliers_of_each_column.sum())
```

```python
# to caluclate each column mean
mean_of_each_column = plotting_columns.mean()

#mean is replacing outliers
for column in plotting_columns.columns:
    outliers_of_each_column = ((plotting_columns[column] < lower_bound[column]) | (plotting_columns[column] > upper_bound[column]))
    plotting_columns.loc[outliers_of_each_column, column] = mean_of_each_column[column]

#to ensure that no outliers are present
outliers_of_each_column = ((plotting_columns < lower_bound) | (plotting_columns > upper_bound)).any(axis=1)
print(outliers_of_each_column.sum())
```

## Statistical tests

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(pd.crosstab(df['Edinburgh_Postnatal_Depression_Scale'], df['Household_Income']))

# Null hypothesis: There is no association between EPDS scores and household income
# Alternate hypothesis: There is an association between EPDS scores and household income

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)
print()

# Determine significance
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant association between EPDS scores and household income.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between EPDS scores and household income.")
```

```python
from scipy.stats import mannwhitneyu
import itertools

# Mapping of numeric values to the descriptive category names
epds_mapping = {
    0: 'None or minimal depression',
    1: 'Mild depression',
    2: 'Moderate depression',
    3: 'Severe depression'
}

# List of unique EPDS categories (numeric values)
epds_categories = df['Edinburgh_Postnatal_Depression_Scale'].unique()

# Generate all possible pairs of categories
pairs = list(itertools.combinations(epds_categories, 2))

# Perform Mann-Whitney U Test for each pair of EPDS categories
for group1_num, group2_num in pairs:
    # Map numeric labels to category names
    group1_name = epds_mapping[group1_num]
    group2_name = epds_mapping[group2_num]
    
    # Get Maternal_Age data for each group
    group1 = df[df['Edinburgh_Postnatal_Depression_Scale'] == group1_num]['Maternal_Age']
    group2 = df[df['Edinburgh_Postnatal_Depression_Scale'] == group2_num]['Maternal_Age']
    
    # Perform Mann-Whitney U test
    stat, p_value = mannwhitneyu(group1, group2)
    
    # Print the result for each comparison using category names
    print(f"Comparison between '{group1_name}' and '{group2_name}':")
    print(f"U-statistic: {stat}, p-value: {p_value}")
    
    # Interpret the p-value
    alpha = 0.05  # significance level
    if p_value < alpha:
        print(f"Since the p-value ({p_value}) is less than {alpha}, we reject the null hypothesis.")
        print("Therefore, there is a significant difference in maternal age between these two EPDS categories.\n")
    else:
        print(f"Since the p-value ({p_value}) is not less than {alpha}, we fail to reject the null hypothesis.")
        print("Therefore, there is no significant difference in maternal age between these two EPDS categories.\n")
```

```python
import pandas as pd
from scipy.stats import spearmanr

Maternal_Age = df['Maternal_Age']
EPDS = df['Edinburgh_Postnatal_Depression_Scale']

# Calculate Spearman's rank correlation coefficient
correlation, p_value = spearmanr(EPDS, Maternal_Age)

print("Spearman's rank correlation coefficient:", correlation)
print("p-value:", p_value)

# Null hypothesis
print("Null Hypothesis (H0): There is no significant correlation between the EPDS and Maternal_Age variables in the population.")

# Alternate hypothesis
print("Alternate Hypothesis (H1 or Ha): There is a significant correlation between the EPDS and Maternal_Age variables in the population.")

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Since the p-value is less than", alpha, ", we reject the null hypothesis.")
    print("Therefore, there is sufficient evidence to conclude that there is a significant correlation between the EPDS and Maternal_Age variables.")
else:
    print("Since the p-value is not less than", alpha, ", we fail to reject the null hypothesis.")
    print("Therefore, there is not enough evidence to conclude that there is a significant correlation between the EPDS and Maternal_Age variables.")
```

```python
import numpy as np
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Define models
rf_classifier = RandomForestClassifier()
knn_model = KNeighborsClassifier(n_neighbors=5)
# Fit models
rf_classifier.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
# Predict probabilities on the test data
y_score_rf = rf_classifier.predict_proba(X_test)
y_score_knn = knn_model.predict_proba(X_test)
# Compute ROC curve and ROC area for each class
n_classes = len(np.unique(y_train))
y_test_bin = label_binarize(y_test, classes=np.unique(y_train))
plt.figure(figsize=(10, 8))
# Set colors for each class
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'green', 'red', 'purple', 'yellow', 'pink']) 
# Plot ROC curves for Random Forest and KNN
for i, color in zip(range(n_classes), colors):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score_rf[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=color, lw=2, label='Random Forest (Class {0}) (AUC = {1:0.2f})'.format(i, roc_auc))
for i, color in zip(range(n_classes), colors):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score_knn[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=color, linestyle='--', lw=2, label='KNN (Class {0}) (AUC = {1:0.2f})'.format(i, roc_auc))

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc="lower right")
plt.show()
```
